# 2. What is the easternmost airport in the United States?

```python
northports = airports[(airports.lon <= 0) & (airports.tzone == 'America/New_York')]
northports.sort_values('lon')
```

According to the given dataset, Eastport Municipal Airport (EPM) is the easternmost airport. All eastern airports will be within the New York timezone, so it's best to filter those only. Finally, since the U.S. is located within the Western Hemisphere, the easternmost airport will be whichever one has the longitude closest to 0, where the Prime Meridian is located. Therefore EPM is the easternmost airport.
